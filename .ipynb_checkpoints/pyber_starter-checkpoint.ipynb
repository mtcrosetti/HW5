```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data_to_load)
ride_df = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
city_ride_df = pd.merge(ride_df, city_df, how="left", on='city')

# Display the data table for preview
city_ride_df.tail()
```

```python
#create DFs for each type
urban_df = city_ride_df.loc[city_ride_df["type"] == "Urban"]
suburban_df = city_ride_df.loc[city_ride_df["type"] == "Suburban"]
rural_df = city_ride_df.loc[city_ride_df["type"] == "Rural"]

u_ride_df = urban_df.groupby(["city"]).count()["ride_id"]
u_fare_df = urban_df.groupby(["city"])["fare"].mean()
u_tot_df = urban_df.groupby("city")["driver_count"].mean()

s_ride_df = suburban_df.groupby(["city"]).count()["ride_id"]
s_fare_df = suburban_df.groupby(["city"])["fare"].mean()
s_tot_df = suburban_df.groupby("city")["driver_count"].mean()

r_ride_df = rural_df.groupby(["city"]).count()["ride_id"]
r_fare_df = rural_df.groupby(["city"])["fare"].mean()
r_tot_df = rural_df.groupby("city")["driver_count"].mean()
```

## Bubble Plot of Ride Sharing Data

```python
# Build the scatter plots for each city types
plt.title("Pyber Ride-Sharing Data")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
types = ["Urban", "Suburban", "Rural"]
# Incorporate the other graph properties
u_s = 10 * (u_tot_df)
s_s = 10 * (s_tot_df)
r_s = 10 * (r_tot_df)
plt.scatter(u_ride_df, u_fare_df, s=u_s, facecolors="orange", edgecolors="black", alpha=0.75)
plt.scatter(s_ride_df, s_fare_df, s=s_s, facecolors="blue", edgecolors="black", alpha=0.75)
plt.scatter(r_ride_df, r_fare_df, s=r_s, facecolors="yellow", edgecolors="black", alpha=0.75)
plt.grid()
#Create a legend
plt.legend(types, loc="best")
#Incorporate a text label regarding circle size
plt.text(5.5,47.5,"Note: Circle sizes correlates with driver count per city.")
#Save Figure
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
urban_fares = urban_df["fare"].sum()
suburban_fares = suburban_df["fare"].sum()
rural_fares = rural_df["fare"].sum()
total_fares = urban_fares + suburban_fares + rural_fares
print(urban_fares)
print(suburban_fares)
print(rural_fares)
u_f_p = (urban_fares / total_fares) * 100
s_f_p = (suburban_fares / total_fares) * 100
r_f_p = (rural_fares / total_fares) * 100

amount = [u_f_p, s_f_p, r_f_p]
labels = ["Urban", "Suburban", "Rural"]
fare_ex = (0.1,0,0)
fare_co = ["r", "b", "y"]

# Build Pie Chart
plt.pie(amount, fare_ex, labels, fare_co, autopct='%.2f', shadow=True, startangle=300)
plt.axis("equal")
# Save Figure
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
urban_rides = urban_df["ride_id"].sum()
suburban_rides = suburban_df["ride_id"].sum()
rural_rides = rural_df["ride_id"].sum()
total_rides = urban_rides + suburban_rides + rural_rides

u_r_p = (urban_rides / total_rides) * 100
s_r_p = (suburban_rides / total_rides) * 100
r_r_p = (rural_rides / total_rides) * 100

r_amount = [u_r_p, s_r_p, r_r_p]
r_labels = ["Urban", "Suburban", "Rural"]
ride_ex = (0.1,0,0)
ride_co = ["r", "b", "y"]
# Build Pie Chart
plt.pie(r_amount, ride_ex, r_labels, ride_co, autopct='%.2f', shadow=True, startangle=300)
plt.axis("equal")
# Save Figure
```

## Total Drivers by City Type

```python
# Calculate Ride Percents
urban_df1 = city_ride_dedup.loc[city_ride_df["type"] == "Urban"]
suburban_df1 = city_ride_dedup.loc[city_ride_df["type"] == "Suburban"]
rural_df1 = city_ride_dedup.loc[city_ride_df["type"] == "Rural"]

urban_drivers = urban_df1["driver_count"].count()
suburban_drivers = suburban_df1["driver_count"].count()
rural_drivers = rural_df1["driver_count"].count()
total_drivers = urban_drivers + suburban_drivers + rural_drivers

u_d_p = (urban_drivers / total_drivers) * 100
s_d_p = (suburban_drivers / total_drivers) * 100
r_d_p = (rural_drivers / total_drivers) * 100

d_amount = [u_d_p, s_d_p, r_d_p]
d_labels = ["Urban", "Suburban", "Rural"]
drive_ex = (0.1,0,0)
drive_co = ["r", "b", "y"]
# Build Pie Chart
plt.pie(d_amount, drive_ex, d_labels, drive_co, autopct='%.2f', shadow=True, startangle=300)
plt.axis("equal")
# Save Figure
```
